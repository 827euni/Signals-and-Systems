```python
from scipy.io import wavfile
from matplotlib import pyplot as plt
import numpy as np
from IPython.display import Audio
```

```python
wav = 'Desktop\\mp3_filename.wav'
d = wavfile.read(wav)
```

```python
print(f"sampling rate: {d[0]} Hz")
```

```python
mywav = d[1]
print(mywav.shape)
```

```python
t = np.arange(0,len(mywav))/d[0]
plt.plot(t,mywav)
plt.xlabel('time (sec)')
plt.ylabel('signal intensity')
plt.grid('on')
plt.show()
```
